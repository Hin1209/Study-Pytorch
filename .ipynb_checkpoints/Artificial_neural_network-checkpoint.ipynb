# 인공 신경망(Artificial Neural Network)   
생물학적 신경망에서 영감을 얻어 만들어졌으며, 크게 입력층, 출력층, 그리고 은닉층으로 구분이된다.   
앞에서 배운 선형 변환과는 달리 인공 신경망에서는 활성화 함수를 이용하여 비선형 변환을 하게된다.  
## 활성화 함수 
활성화 함수에는 여러 종류가 있는데 그 중 몇가지만 살펴보면  
- sign(signum)
- Linear
- Piece-wise-linear
- Logistic(sigmoid)
- Hyperbolic tangent
- Rectifier, ReLU(Rectified Linear Unit) 이 있다. 
   
## 전파와 역전파   

전파 : 입력값이 들어오면 여러 개의 은닉층을 순서대로 거쳐 결과값을 내는 과정을 전파라고 한다.(forward propagation)  
역전파 : 결과와 정답의 차이로 계산된 손실을 연쇄법칙(chain rule)을 이용하여 입력 단까지 전달하는 과정을 역전파라고 한다.(back propagation)   


```python
import torch
import torch.nn as nn 
import torch.optim as optim 
import torch.nn.init as init
import matplotlib.pyplot as plt

num_data = 1000
num_epoch = 10000

noise = init.normal_(torch.FloatTensor(num_data, 1), std = 1)
x = init.uniform_(torch.Tensor(num_data, 1), -15, 15)
y = (x**2) + 3
y_noise = y + noise

# Sequential 클래스를 이용해 입력값이 들어오면 입력을 받은 모듈을 순서대로 실행하여 결과값을 리턴한다.  
model = nn.Sequential(  
        nn.Linear(1, 6),
        nn.ReLU(),
        nn.Linear(6, 10),
        nn.ReLU(),
        nn.Linear(10, 6),
        nn.ReLU(),
        nn.Linear(6, 1),
)

loss_func = nn.L1Loss()
optimizer = optim.SGD(model.parameters(), lr = 0.0002)

loss_array = []
for i in range(num_epoch):
  optimizer.zero_grad()
  output = model(x)
  loss = loss_func(output, y_noise)
  loss.backward()
  optimizer.step()

  loss_array.append(loss)

plt.plot(loss_array)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x.numpy(),y_noise,label="Original Data")
plt.scatter(x.detach().numpy(),output.detach().numpy(),label="Model Output")
plt.legend()
plt.show()
```
